```python
# importing package

import matplotlib.pyplot as plt
import numpy as np
```

```python
# x_axis data 
x = [1,2,3,4,5]

# RFC Accuracy's
RFC = [73, 77, 76, 75, 73]

# SVM Accuracy's
SVM = [83, 80, 81, 79, 81]
```

```python
# plot lines

plt.plot(x, RFC, label = "RFC", linestyle="-")
plt.plot(x, SVM, label = "SVM", linestyle="--")
plt.title("Comparison pf RFC vs SVM over a period of 5 times")
plt.xlabel('Period over 5 times')
plt.ylabel('Accuracy of Classifier')
plt.legend()
plt.show()
```
